
# KA-PO Exponential model (log loss) comparisons

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import sys
sys.path.append('..')
from paths import*
```

```python
paths = Paths()
fig_save_path = paths.fig_save_path + 'exponential_model/'
data = pd.read_excel(paths.model_outputs +'/model_params.xlsx', header = [1])
```

```python
log_loss_df = data[[ 'Monkey', 'Model name ', 'log loss']]
log_loss_df.loc[:, 'model_name'] = log_loss_df.loc[:, 'Model name '].str.replace('KA_', '').str.replace('PO_', '')
```

```python
log_loss_df = log_loss_df.pivot(index = 'model_name', columns=['Monkey'], values='log loss')
```

```python
fig, ax = plt.subplots(1,1, facecolor = 'white')


sns_plot = sns.heatmap(log_loss_df, annot=True, fmt = '.3f')
sns_plot.set_ylabel('Model name', fontsize = '14')
sns_plot.set_xlabel('Monkey', fontsize = '14');
sns_plot.set_title('Log losses for different models', fontsize = 16, y = 1.05)

#fig = sns_plot.get_figure()
fig.savefig(fig_save_path + "log_losses.png", dpi = 300, bbox_inches='tight')
```
